# Variance

## Definition
Variance measures how much the data points differ from the mean (average). It shows how spread out the values are.

## Formula
$$\text{Variance} = \frac{1}{N} \times \sum_{i=1}^{N} (x_i - \mu)^2$$

### Where:
- $x_i$ = each data point
- $\mu$ = mean (average) of the data
- $N$ = total number of data points

## Alternative Notation
```
Variance = (1 / N) × Σ (xi – μ)²
```

## Example
Given data: 2, 4, 6, 8, 10

1. Calculate mean: μ = (2+4+6+8+10)/5 = 6
2. Calculate squared differences:
   - (2-6)² = 16
   - (4-6)² = 4
   - (6-6)² = 0
   - (8-6)² = 4
   - (10-6)² = 16
3. Variance = (16+4+0+4+16)/5 = 8


````python
data = [2, 4, 4, 4, 5, 5, 7, 9]

mean = sum(data) / len(data)
print("Mean : ", mean)

variance = sum((x - mean)**2 for x in data) / len(data)
print("Variance : ", variance)
````
